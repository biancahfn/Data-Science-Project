## Introduction ##

- #### Background Information: ####

Cardiovascular diseases (CVDs) are a significant cause of mortality globally, often accompanied by symptoms such as chest pain, which can indicate underlying cardiac issues. Understanding the correlation between various health metrics and chest pain types is critical for accurate diagnosis and effective treatment planning.

- #### Question ####
Can we predict the type of chest pain experienced by a patient based on their cardiovascular health metrics such as cholesterol levels, and maximum heart rate achieved?

### Reasoning Behind Predictors ###
- Cholesterol Levels (Cholestrol)
Cholesterol is a crucial factor in cardiovascular health. High cholesterol levels are a significant risk factor for coronary artery disease, which can lead to various types of chest pain, particularly angina.
- Maximum Heart Rate Achieved
The maximum heart rate achieved during an exercise test is a measure of cardiac function and fitness. It can indicate how well the heart copes with increased workload and can reflect underlying heart conditions.


## Dataset Information ##

- #### Demographic Information: ####

- `Age`: Age of the patient.
- `Sex`: Sex of the patient (1 = male, 0 = female).

- #### Cardiovascular Health Metrics: ####


- `Exercise Induced Angina (exang)`: Indicates whether the patient experiences exercise-induced angina (1 = yes, 0 = no).
- `Number of Major Vessels (ca)`: Represents the number of major vessels observed via diagnostic procedures (0-3).
- `Resting Blood Pressure (trtbps)`: Measures the resting blood pressure of the patient (in mm Hg).
- `Cholesterol Levels (chol)`: Represents the cholesterol levels of the patient (in mg/dl).
- `Fasting Blood Sugar (fbs)`: Indicates whether the patient has fasting blood sugar levels > 120 mg/dl (1 = true, 0 = false).
- `Resting Electrocardiographic Results (rest_ecg)`: Represents the resting electrocardiographic results of the patient (0 = -  normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria).
- `Maximum Heart Rate Achieved (thalach)`: Represents the maximum heart rate achieved during exercise.

- #### Chest Pain Types: ####

`Chest Pain Type (cp)`: Represents the type of chest pain experienced by the patient (1 = typical angina, 2 = atypical angina, 3 = non-anginal pain, 4 = asymptomatic).

Understanding the relationship between these health metrics and chest pain types can aid healthcare professionals in diagnosing CVDs accurately and developing personalized treatment plans. By analyzing these relationships, this study aims to contribute to improved patient care and outcomes in the context of cardiovascular health.

- Make the chest pain column a category variable type as that will be our classifer.

```r


colnames(heart_attack) <- c('Age', 'Sex', 'Chest_Pain_Type', 'Cholestrol', 'Resting_blood_pressure', 'Fasting_Blood_Sugar', 'Resting_electrocardiographic_results',
                            'Maximum_heart_rate_achieved', 'Exercise_induced_angina', 'Previous_Peak', 'Slope', 'Number_Of_Vessels', 'Thal_Rate', 'Target_Variable')



heart_attack
```

Now we will get rid of the variables that won't be needed for the analysis and make Chest_Pain_Type a categorical variable and include the actual type of chest pains instead of numerical values.

```r
heart_attack <- select(heart_attack, Chest_Pain_Type, Cholestrol, Maximum_heart_rate_achieved) |>
                mutate(Chest_Pain_Type = as.factor(Chest_Pain_Type)) |>
                mutate(Chest_Pain_Type = fct_recode(Chest_Pain_Type,
                                                    "Typical Angina" = "0",
                                                    "Atypical Angina" = "1",
                                                    "Non-anginal Pain" = "2",
                                                    "Asymptomatic" = "3"))
                                                    

heart_attack
```

### Splitting the dataset ###

Now that we have done the cleaning and wrangling to the dataset, it is time to split the data into the traning and testing datasets which will help us in the classification process which will be explained in the methodology section. We will use initial_split to split our dataframe into 75% training and 25% testing, since it shuffles our data for us and ensures an constant proportion of each class is present in both. The 75-25 split allows us to train our model on as many data points as possible while also keeping enough data for effective testing later.

```r
split <- initial_split(heart_attack, prop = 0.75, strata = Chest_Pain_Type)
heart_attack_training <- training(split)
heart_attack_testing <- testing(split)

heart_attack_training
heart_attack_testing
```

### Beyond this visualization ###
k-nearest neighbors (k-NN) model can help analyze the relationship between cholesterol levels, maximum heart rate, and chest pain types by considering the proximity of data points in a multidimensional space, potentially uncovering patterns not visible in a simple scatter plot. By comparing a patient's data to that of their 'nearest neighbors' in terms of similar health metrics, k-NN can predict the type of chest pain, even in the absence of a clear linear relationship

## Tuning to find the best value for K ##

```r
set.seed(1)

# Performing Standardization 
heart_recipe <- recipe(Chest_Pain_Type ~ Cholestrol + Maximum_heart_rate_achieved , data = heart_attack_training) |>
   step_scale(all_predictors()) %>%
   step_center(all_predictors())

# K-nearest neighbour model specification
knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
            set_engine("kknn") |>
            set_mode("classification")

# Perform 10 fold-cross valdiation
heart_vfold <- vfold_cv(heart_attack_training, v = 10, strata = Chest_Pain_Type)

gridvals <- tibble(neighbors = c(1:100))

# Create workflow
knn_results <- workflow() |>
      add_recipe(heart_recipe) |>
      add_model(knn_tune) |>
      tune_grid(resamples = heart_vfold, grid = gridvals) |>
# Collect metrics, filter for accuracy
      collect_metrics() |>
      filter(.metric == "accuracy") |>
      select(neighbors, mean) |>
      arrange(desc(mean))

head(knn_results)
```

Looks like $K=75$, $K=74$ and $K=76$ all give an accuracy around 49.5% hence we need to graph an accuracy vs #neighbours graph.

```r
accuracy_versus_k <- ggplot(knn_results, aes(x = neighbors, y = mean))+
      #geom_point() +
      geom_line() +
      labs(x = "Neighbors", y = "Accuracy Estimate", title = "Accuracies for 1 ≤ k ≤ 100") +
      scale_x_continuous(breaks = seq(0, 100, by = 5)) + # adjusting the x-axis
      theme(text = element_text(size = 15))

accuracy_versus_k
```

Looks like $K=75$ is the value to go for in the analysis.

## KNN Model ##
Note: $K=75$

```r
set.seed(1)

final_spec <- nearest_neighbor(weight_func="rectangular", neighbors = 75) |>
    set_engine("kknn") |>
    set_mode("classification")

fit <- workflow() |>
    add_recipe(heart_recipe) |>
    add_model(final_spec) |>
    fit(data = heart_attack_training)

fit
```

## Accuracy and Classifier analysis ##

Now, we will finally use the testing set to look at how our classifier performs when given data it has never seen before.

To do this, we will predict using the same fit as before, but this time with heart_attack_testing. We need to look at its accuracy using metrics and its confusion matrix using conf_mat.

```r
set.seed(1)

#predict
predictions <- predict(fit, heart_attack_testing) |>
               bind_cols(heart_attack_testing)

#find the accuracy of the predictions
metrics<- predictions %>%
    metrics(truth = Chest_Pain_Type, estimate = .pred_class) |>
    filter(.metric == "accuracy") 

metrics

#confusion matrix, identifies the areas the model is weak at classifying
conf_mat<- predictions |>
    conf_mat(truth = Chest_Pain_Type, estimate = .pred_class)
conf_mat
```

The k-nearest neighbors (k-NN) model in question has achieved an accuracy of approximately 51.9%, as indicated by the output. This figure represents the ratio of correct predictions to the total number of predictions made on the testing dataset. An accuracy rate above 50% suggests that the model is performing better than random chance, but it also indicates significant room for improvement, as nearly half of the predictions are incorrect.


The confusion matrix provides a detailed view of the model's performance across the different classes of chest pain. Correct predictions are found along the diagonal, where the predicted class aligns with the true class. For instance, 'Typical Angina' was correctly predicted 33 times. However, the confusion matrix also reveals instances of misclassification, such as 'Typical Angina' being incorrectly predicted as 'Non-anginal Pain' 15 times.


The absence of correct predictions for the 'Asymptomatic' class suggests that the model may struggle with classes that have fewer instances or are less well-defined by the features used. This could be indicative of class imbalance or insufficient discriminative information within the feature set to distinguish this class from others.
